# Plot Results

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
HIST_RAND_FILENAME = "rand_results_SIMPLE.csv"
HIST_MCTS_FILENAME = "mcts_results_SIMPLE.csv"
```

```python
df_rand = pd.read_csv(HIST_RAND_FILENAME) 
df_mcts = pd.read_csv(HIST_MCTS_FILENAME) 
```

```python
def plot_soc(df): 
    plt.figure(figsize=[8,4]) 
    plt.plot(df["t"], df["soc"], label="SOC", alpha=0.5)
    plt.plot(df["t"], df["tou"], label="TOU", alpha=0.5) 
    plt.xlabel("Time [TOD 1-5]")
    plt.legend() 
    plt.show() 
    return

def plot_temp(df): 
    plt.figure(figsize=[8,4]) 
    plt.plot(df["t"], df["odt"], label="ODT", alpha=0.5)
    plt.plot(df["t"], df["rmt"], label="RMT", alpha=0.5) 
    plt.plot(df["t"], df["hsp"], label="HSP", alpha=0.5) 
    plt.plot(df["t"], df["csp"], label="CSP", alpha=0.5) 
    plt.plot(df["t"], df["occ"], label="OCC", alpha=0.5) 
    plt.xlabel("Time [TOD 1-5]")
    plt.ylabel("Temp [degF]") 
    plt.legend() 
    return 

def plot_load(df): 
    plt.figure(figsize=[8,4])
    plt.plot(df["t"], df["d_"], label="Non-HVAC", alpha=0.6)
    plt.plot(df["t"], df["d_hv"], label="HVAC", alpha=0.6) 
    plt.plot(df["t"], df["tou"], label="TOU", alpha=0.6) 
    plt.plot(df["t"], df["odt"], label="ODT", alpha=0.6) 
    plt.xlabel("Time [TOD 1-5]")
    plt.legend()
    return 

```

```python
plot_soc(df_mcts)
```

```python
plot_load(df_mcts) 
```

```python
plot_temp(df_mcts)
```

```python
plot_soc(df_rand) 
```

```python
plot_temp(df_rand) 
```
